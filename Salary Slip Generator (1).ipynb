```python
#text = "Father Name\nIncome Tax Number (PAN):\nUniversal Account Number (UAN)\nBank Details:\nAccount Number:\nMode of Payment\nLeave"


# Python program to create
# a pdf file


from fpdf import FPDF
import pandas as pd
import numpy as np

data = pd.read_excel("sample_excel.xlsx")

for i in range(data.shape[0]):
    # save FPDF() class into a
    # variable pdf
    pdf = FPDF()

    # Add a page
    pdf.add_page()

    # set style and size of font
    # that you want in the pdf
    pdf.set_font("Arial", size = 8)
    pdf.rect(10, 45, 185, 160, style='')

    #pdf.set_xy(206, 661)
    pdf.text(70, 42, "SOURCE SOFT SOLUTIONS PRIVATE LIMITED")

    pdf.text(98, 49, "Pay Slip")
    pdf.text(95, 53, "for April-2022")

    pdf.line(10, 56, 195 , 56)

    pdf.text(95, 60, data.iloc[i]["Name of Employee"])

    pdf.text(12, 65, "Employee Number:")
    pdf.text(58, 65, data.iloc[i]["Emp Code"])
    pdf.text(105, 65, "Father Name")
    pdf.text(160, 65, data.iloc[i]["FATHERS NAME"])

    pdf.text(12, 70, "Department")
    pdf.text(58, 70, "US Staffing")
    pdf.text(105, 70, "Income Tax Number (PAN):")
    pdf.text(160, 70, data.iloc[i]["PAN NO."])

    pdf.text(12, 75, "Designation:")
    pdf.text(58, 75, data.iloc[i]["Designation"])
    pdf.text(105, 75, "Universal Account Number (UAN)")
    pdf.text(160, 75, str(data.iloc[i]["UAN NO."]))

    pdf.text(12, 80, "Location:")
    pdf.text(58, 80, data.iloc[i]["Location"])
    pdf.text(105, 80, "Bank Details:")
    pdf.text(160, 80, "ICICI Bank Ltd")

    pdf.text(12, 85, "Date of joining:")
    pdf.text(58, 85, str(data.iloc[i]["DATE OF JOIN"]))
    pdf.text(105, 85, "Account Number:")
    pdf.text(160, 85, str(data.iloc[i]["Bank Account No"]))

    pdf.text(12, 90, "Sex")
    pdf.text(58, 90, data.iloc[i]["SEX"])
    pdf.text(105, 90, "Mode of Payment")
    pdf.text(160, 90, "Bank Transfer")

    pdf.text(12, 95, "Present Days")
    pdf.text(58, 95, str(data.iloc[i]["Payable days"]))
    pdf.text(105, 95, "Leave")
    pdf.text(160, 95, str(data.iloc[i]["Deduction"]))

    pdf.line(10, 100, 195 , 100)

    pdf.text(12, 105, "Earnings")
    pdf.text(48, 105, "Amount")
    pdf.text(75, 105, "Gross Salary")
    pdf.text(115, 105, "Deductions")
    pdf.text(150, 105, "Amount")
    pdf.text(178, 105, "Gross Salary")

    pdf.line(10, 110, 195 , 110)
    pdf.line(10, 138, 195 , 138)

    pdf.text(12, 114, "Basic")
    pdf.text(12, 118, "House Rent Allowance")
    pdf.text(12, 122, "Conveyance Allowance")
    pdf.text(12, 126, "Medical Allowance")
    pdf.text(12, 130, "Special Allowance")
    pdf.text(12, 134, "Arrear")
    pdf.line(45, 100, 45, 145)

    if np.isnan(data.iloc[i]["Basic"]):
        basic = 0.00
    else:
        basic = data.iloc[i]["Basic"]
        
    if np.isnan(data.iloc[i]["HRA"]):
        hra = 0.00
    else:
        hra = data.iloc[i]["HRA"]
        
    if np.isnan(data.iloc[i]["Conveyance Allowance"]):
        conv_allowance = 0.00
    else:
        conv_allowance = data.iloc[i]["Conveyance Allowance"]
        
    if np.isnan(data.iloc[i]["Special allowance"]):
        spec_allowance = 0.00
    else:
        spec_allowance = data.iloc[i]["Special allowance"]     

    if np.isnan(data.iloc[i]["Arrears"]):
        arrears = 0.00
    else:
        arrears = data.iloc[i]["Arrears"]     
    
    pdf.text(48, 114, str("{:.2f}".format(basic)).rjust(25))
    pdf.text(48, 118, str("{:.2f}".format(hra)).rjust(25))
    pdf.text(48, 122, str("{:.2f}".format(conv_allowance)).rjust(25))
    pdf.text(48, 126, "0.00".rjust(25))
    pdf.text(48, 130, str("{:.2f}".format(spec_allowance)).rjust(25))
    pdf.text(48, 134, str("{:.2f}".format(arrears)).rjust(25))
    pdf.line(74, 100, 74, 145)

    pdf.text(80, 114, "42050.00".rjust(19))
    pdf.text(80, 118, "16820.00".rjust(19))
    pdf.text(80, 122, "1547.00".rjust(19))
    pdf.text(80, 126, "0.00".rjust(19))
    pdf.text(80, 130, "23683.00".rjust(19))
    pdf.text(80, 134, "0.00".rjust(19))
    pdf.line(101, 100, 101, 153)

    pdf.text(103, 114, "Employee's PF Contribution".ljust(50))
    pdf.text(103, 118, "Employee's ESI Contribution".ljust(50))
    pdf.text(103, 122, "TDS".ljust(50))
    pdf.text(103, 126, "Loan/Advance".ljust(50))
    pdf.text(103, 130, "Bonus".ljust(50))
    pdf.line(145, 100, 145, 153)
    
    if np.isnan(data.iloc[i]["PF Employee"]):
        pf_contribution = 0.00
    else:
        pf_contribution = data.iloc[i]["PF Employee"]
        
    if np.isnan(data.iloc[i]["PF Employer"]):
        esi_contribution = 0.00
    else:
        esi_contribution = data.iloc[i]["PF Employer"]
        
    if np.isnan(data.iloc[i]["TDS"]):
        tds = 0.00
    else:
        tds = data.iloc[i]["TDS"]
        
    if np.isnan(data.iloc[i]["Advance\nDeductions"]):
        advance_deduc = 0.00
    else:
        advance_deduc = data.iloc[i]["Advance\nDeductions"]     
        
    pdf.text(142, 114, str("{:.2f}".format(pf_contribution)).rjust(28))
    pdf.text(142, 118, str("{:.2f}".format(esi_contribution)).rjust(28))
    pdf.text(142, 122, str("{:.2f}".format(tds)).rjust(28))
    pdf.text(142, 126, str("{:.2f}".format(advance_deduc)).rjust(28))
    pdf.text(142, 130, "0.00".rjust(28))
    pdf.line(170, 100, 170, 153)

    pdf.text(178, 114, "1850.00".rjust(15))
    pdf.text(178, 118, "0.00".rjust(15))
    pdf.text(178, 122, "0.00".rjust(15))
    pdf.text(178, 126, "0.00".rjust(15))
    pdf.text(178, 130, "0.00".rjust(15))
    #pdf.line(1, 100, 175, 138)

    pdf.line(10, 145, 195, 145)
    
    pdf.text(12, 142, "Total Earnings")
    total_earnings = basic + hra + conv_allowance + spec_allowance + arrears
    pdf.text(48, 142, str("{:.2f}".format(total_earnings)).rjust(25))
    
    pdf.text(80, 142, "87000.00".rjust(19))
    pdf.text(103, 142, "Total Deductions".ljust(50))
    total_deductions = pf_contribution + esi_contribution + tds + advance_deduc
    pdf.text(142, 142, str("{:.2f}".format(total_deductions)).rjust(28))
    pdf.text(178, 142, "1850.00".rjust(15))

    pdf.line(10, 153, 195, 153)
    pdf.text(103, 150, "Net Amount".ljust(50))
    pdf.text(176, 150, ("Rs. " + str("{:.2f}".format(total_earnings - total_deductions))).rjust(12))

    pdf.text(50, 203, "This is a computer generated document and does not require any signatures.")

    # Amount Gross Salary Deductions Amount Gross Salary
    # save the pdf with name .pdf
    pdf.output(data.iloc[i]["Name of Employee"] + ".pdf")
    
print("Done!!!")
```
